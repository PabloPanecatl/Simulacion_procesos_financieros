```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def funcion_a_integrar(x):
    return x**2

a = 2
b = 5

N = 100

x = np.random.uniform(a, b, N)
y = np.random.uniform(0, funcion_a_integrar(b), N)
```

```python
n_bajo_curva = np.sum(y < funcion_a_integrar(x))
n_bajo_curva
```

```python
respuesta = (b-a) * funcion_a_integrar(b) * (n_bajo_curva / N)
respuesta
```

```python
plt.plot(x,y,'.')
plt.plot(np.linspace(a,b,100),funcion_a_integrar(np.linspace(a,b,100)),'r')
plt.show()
```

# Actividad de clase:
Calcular Pi

```python
def funcion_circulo(x, y):
    return x**2 + y**2

r = 1
N = 10000000

area_cuadrado = 4
x = np.random.uniform(-r, r, N)
y = np.random.uniform(-r, r, N)

puntos_dentro = np.sum(funcion_circulo(x, y) <= r**2)
estimacion_area_circulo = area_cuadrado * puntos_dentro / N
estimacion_area_circulo
```

```python
# Graficamos los puntos y el circulo
theta = np.linspace(0, 2*np.pi, 100)
plt.figure(figsize=(6,6))
plt.plot(r * np.cos(theta), r * np.sin(theta), 'r')  # Crculo
plt.scatter(x, y, s=1)
plt.xlim(-r, r)
plt.ylim(-r, r)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
